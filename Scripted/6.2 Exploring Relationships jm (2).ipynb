# 5.Categorical plots

```python
# creating a histogram to see the price of houses most sold
sns.histplot(df['price'],bins=75,kde=True)
```

##### Houses on the lower side of the price are sold at a much higher frequency than more expensive houses. 

```python
sns.histplot(df['grade'],bins=75,kde=True)
```

##### • grade - An index from 1 to 13, where 1-3 falls short of building construction and design, 7 has an average level of construction and design, and 11-13 have a high quality level of construction and design. The average contruction and design sells more houses than any other grade.

```python
sns.histplot(df['view'],bins=75,kde=True)
```

##### view - An index from 0 to 4 of how good the view of the property was.
This shows that the view of the propery doesn't really make a difference in sales. The highest sales are with a view rated 0.

```python
sns.histplot(df['yr_renovated'],bins=75,kde=True)
```

##### Renevation in the house before selling does not impact the sales of the house.

```python
sns.histplot(df['waterfront'],bins=75,kde=True)
```

##### Having a waterfront property does not have any affect on the sales of a house. As the houses that were sold the most didn't have any waterfront.

```python
df.loc[df['price'] < 200000, 'price category'] = 'Low price'
df.loc[(df['price'] >= 200000) & (df['price'] < 500000), 'price category'] = 'Middle price'
df.loc[df['price'] >= 500000, 'price category'] = 'High price'
```

```python
df['price category'].value_counts(dropna=False)
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g= sns.catplot(x="price category", y="sqft_living", hue="price category", data=df)
```
